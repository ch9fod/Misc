```python
#%matplotlib
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
x, a, m = symbols('x a m')
init_printing()
# data
xs = np.linspace(0, 2, 11)
ys = [0, 2.40, 3.00, 2.55, 2.24, 1.72, 1.18, 0.82, 0.56, 0.42, 0.25]
print('x =', xs)
print('y =', ys)
```

```python
# equation
eq = a * x * exp(x*m); eq
```

```python
# summation of squared values for:
# yi - f(xi)
summ = 0
for i in range(len(ys)):
    summ = summ + (ys[i] - eq.subs(x, xs[i]))**2
summ
```

```python
# partial derivative with respect to a
f1 = diff(summ, a); f1
```

```python
# partial derivative with respect to m
f2 = diff(summ, m); f2
```

```python
# solving and calculating a and m
# using a non-linear solver with initial values
av, mv = nsolve((f1, f2), (a, m), (8, 0), verify=False);
av = float(av)
mv = float(mv)
print ('a = {:0.3f}'.format(av))
print ('m = {:0.3f}'.format(mv))
```

```python
newy = []
for i in range(len(xs)):
    newy.append(eq.subs([(x, xs[i]), (a, av), (m, mv)]))
```

```python
# plot values
fig = plt.figure(figsize=(12,9))
ax = fig.add_subplot(111)
ax.plot(xs, ys, 'ro', label='Original Data')
ax.plot(xs, newy, 'b-', label='fit')
ax.text(1.5, 2.5, "a = {:0.3f}".format(av), fontsize=15)
ax.text(1.5, 2.4, "m = {:0.3f}".format(mv), fontsize=15)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
